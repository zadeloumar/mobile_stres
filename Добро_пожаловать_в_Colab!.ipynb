# Дорожная карта развития: Fullstack → Backend + DevOps

**Цель:** перейти от базового fullstack к уверенным навыкам серверной разработки и практик DevOps. Backend-разработка включает создание серверной логики, работу с базами данных и API. DevOps – это набор практик, объединяющих команды разработки и эксплуатации для более быстрой и качественной поставки ПО. Дальнейший план разбит на месяцы. Ежедневно уделяйте ≈3 часа (например, 3 микросессии по 1 часу), используя короткие уроки и практику (видео по 20–30 мин, задачи на Codewars/FreeCodeCamp и т.п.) для закрепления материала.

## Общие рекомендации

* **Распределение времени:** учитесь каждый день по 2–3 часа, разбивая время на микро-сессии (видео-уроки, статьи, кодинг) и мини-практику. Это поможет усваивать материал между рабочими сменами.
* **Практика через проекты:** сразу применяйте знания на реальных задачах (мини-проекты, задачи из онлайн-курсов). Например, при изучении HTTP/API – создайте простой HTTP-сервер; при изучении БД – набросайте таблицы и запросы.
* **Используйте сообщества:** задавайте вопросы на StackOverflow, читайте GitHub – **GitHub** и **StackOverflow** упоминаются как ключевые бесплатные ресурсы для совместной работы и обмена знаниями.
* **Ресурсы:** бесплатные образовательные порталы (MDN Web Docs, W3Schools, freeCodeCamp, Codecademy и др.) содержат подробные руководства по веб-разработке. (Например, MDN – обширный справочник по HTML/CSS/JS, бесплатный курс freeCodeCamp охватывает полный стек).
* **Измерение прогресса:** ставьте конкретные цели по завершению тем (научиться запускать сервер, настраивать окружение, писать скрипты). Проверяйте себя, делая проекты или задачи (см. «Практика»). Регулярно ревьюьте код и сравнивайте с образцами.

## Месяц 1: Основы веба и инструментов

* **Фокус:** изучите базовый веб-стек: как работает Интернет, протокол HTTP, DNS и хостинг. Повторите основы HTML/CSS/JavaScript (для полноты картины). Освойте систему контроля версий Git: коммиты, ветки, слияния – Git используется для контроля версий в разработке.
* **Ресурсы:** MDN Web Docs (HTTP, HTML, CSS, JS); курсы/статьи по HTTP (например, MDN или Cloudflare); руководство **Pro Git** и туториал Atlassian (Git); интерактивные уроки (W3Schools, freeCodeCamp, CodeAcademy).
* **Практика:** создайте простую одностраничную HTML/CSS-страницу и разместите её на GitHub Pages или в локальном сервере. Опубликуйте код в репозитории, отрабатывая `git init`, `add`, `commit`, `push`.
* **Оценка:** вы научитесь развертывать локальный веб-сервер, делать HTTP-запросы и получать ответы. Проверьте себя: запустили ли сайт, корректно ли обновляются файлы, можете ли объяснить работу HTTP и Git.
* **Упор:** на понимании работы веба «под капотом», разнице между frontend и backend, уверенном использовании Git. Эти навыки – фундамент для последующего изучения бекенда.

## Месяц 2: PHP и Laravel – расширение бэкенда

* **Фокус:** углубитесь в PHP и его современные практики. PHP – один из основных языков бэкенда. Изучите ООП в PHP (наследование, интерфейсы, namespace), менеджер пакетов Composer. Начните работать с фреймворком Laravel: маршруты, контроллеры, миграции и ORM (Eloquent).
* **Ресурсы:** официальная документация Laravel (бесплатна); серия видеокурсов Laracasts (имеются бесплатные уроки); статьи и курсы на YouTube (по поиску «Laravel 10 for beginners»). Для PHP – сайт PHP.net, курс PHP на freeCodeCamp.
* **Практика:** разработайте простое CRUD-приложение на Laravel (например, блог или список задач) с регистрацией и авторизацией пользователей (Auth в Laravel). Используйте БД (MySQL/PostgreSQL) для хранения данных (репозиторий пользователей, постов и т.д.).
* **Оценка:** оцените умение работать с Laravel: например, создайте новый роут, модель и миграцию – должно правильно сохраняться в БД. Проверьте, умеете ли отлаживать запросы и просматривать логи. Прогресс виден по количеству реализованных функций (регистрация, CRUD-записи и т.д.).
* **Упор:** на понимании структур MVC и особенностей Laravel: маршрутов, middlewares, систем авто-перезагрузки. Следите за чистотой кода и оформлением (PSR-стандарты). Начните писать простые тесты (например, проверять, что страница загрузилась) – это повысит уверенность в коде.

## Месяц 3: Базы данных и API

* **Фокус:** погружение в реляционные СУБД. Углубите знания SQL: сложные запросы (JOIN, GROUP BY), индексы, нормализация данных. Создайте продвинутые миграции и сиды в Laravel. Изучите принципы REST и создайте API на Laravel (ресурсные контроллеры, аутентификация через API токены).
* **Ресурсы:** курсы SQL на Khan Academy или W3Schools, бесплатные туториалы на YouTube (“SQL Tutorial”). Документация Laravel по Eloquent и API Resources.
* **Практика:** расширьте проект из прошлого месяца: сделайте к нему API-интерфейс – точки доступа для получения/создания/удаления данных в формате JSON. Используйте Postman или Insomnia для тестирования запросов. Добавьте логирование ошибок.
* **Оценка:** прогресс наглядно через работоспособность API: все типичные запросы должны возвращать ожидаемые данные. Проверьте знание SQL через написание сложных запросов к базе (например, сводные отчёты).
* **Упор:** на структурировании данных и оптимизации запросов. Важно уметь объяснить свою модель данных и обосновать архитектуру API. Следите за валидацией данных и сообщениями об ошибках (так работает backend).

## Месяц 4: Расширение навыков бэкенда и основы систем

* **Фокус:** дополните стек: при желании освоите другой язык (например, JavaScript/Node.js или Python) для сравнения и скриптинга. Изучите основы системного администрирования: базовые команды Linux (консоль, навигация, скрипты bash). Работа с командной строкой нужна для любой DevOps-задачи.
* **Ресурсы:** туториалы Linux Basics (например, linuxcommand.org, Codecademy), вводные видео по CLI. Для Node.js – сайт javascript.info или «Express.js crash course» на YouTube, если решите попробовать backend на JS.
* **Практика:** установите и настройте виртуальную машину Linux (или WSL 2). Напишите простой bash-скрипт, который, например, делает резервную копию вашей базы. Если изучали Node.js – создайте простое Express-приложение и разверните его локально.
* **Оценка:** проверяйте себя умением ориентироваться в терминале: создайте папки, файлы, привилегии, выполните `grep`, `awk` и др. Создайте и запустите bash-скрипт без ошибок. Если работали с Node, убедитесь, что можете запустить и остановить сервер.
* **Упор:** на автоматизации рутинных задач и уверенной работе в ОС. Это основа для последующего развертывания сервисов и DevOps – без понимания терминала сложно двигаться дальше.

## Месяц 5: Контейнеризация и DevOps-инструменты

* **Фокус:** знакомство с Docker и виртуализацией. Научитесь писать `Dockerfile`, собирать контейнерные образы и запускать контейнеры. Это ключевой навык для DevOps: упаковывать приложения и их окружение.
* **Ресурсы:** официальная документация Docker (Docker Docs), интерактивные руководства (Play with Docker, Katacoda). Курсы Docker на YouTube (например, “Docker Tutorial for Beginners”). Обратите внимание: Git как система контроля используется не только для кода, но и для инфраструктуры.
* **Практика:** создайте `Dockerfile` для вашего Laravel-приложения (с PHP, Composer, веб-сервером) и запустите приложение внутри контейнера. Научитесь запускать несколько контейнеров через Docker Compose (например, веб-сервер + база).
* **Оценка:** прогресс – приложение успешно работает в контейнере, доступно по локальному порту. Проверьте умение писать `docker build`, `docker run`, использовать Compose (поднимать несколько контейнеров одной командой).
* **Упор:** на понимании, как контейнеры изолируют окружение и упрощают деплой. Следите за тем, чтобы в контейнере были только нужные зависимости (минимизируйте образ). Попрактикуйтесь в том, чтобы в любой момент повторить запуск контейнера по инструкции (Dockerfile).

## Месяц 6: Облачные платформы и развёртывание

* **Фокус:** познакомьтесь с облачными сервисами (AWS/Azure/GCP) или простыми PaaS (Heroku). Цель – научиться разворачивать приложение в облаке.
* **Ресурсы:** бесплатные учебники AWS (AWS Free Tier, бесплатные курсы Cloud Practitioner), Google Cloud или Azure (есть бесплатные кредиты для новичков). Также бесплатные платформы: Heroku имеет бесплатный тариф для небольших приложений.
* **Практика:** выберите платформу (например, AWS EC2 или Heroku) и разверните там своё приложение: настройте сервер (или создайте приложение в Heroku), перенесите код, настройте подключение к базе данных.
* **Оценка:** успех – приложение доступно по интернет-адресу (URL или IP) без ошибок. Проверьте, что вы умеете перезапускать сервис, смотреть логи в облаке.
* **Упор:** на навыках настройки серверов/инстансов и работе с облачными инструментами. Например, на AWS важно понимать конфигурацию безопасности (security groups, SSH-ключи) и управление ресурсами через консоль.

## Месяц 7: CI/CD и автоматизация сборки

* **Фокус:** реализуйте непрерывную интеграцию и доставку (CI/CD). Освойте один из инструментов (GitHub Actions, GitLab CI/CD, Jenkins). Это позволит автоматически тестировать и деплоить приложение при изменениях.
* **Ресурсы:** официальные туториалы по выбранному инструменту (например, GitHub Actions documentation). На YouTube – “GitHub Actions tutorial”, “GitLab CI tutorial”.
* **Практика:** настройте CI-пайплайн: при пуше в GitHub запускаются тесты (или хотя бы простая сборка проекта). Затем настройте CD: по успешному тесту – деплой на сервер.
* **Оценка:** проверьте работу пайплайна: смоделируйте провал и успех (например, измените код и посмотрите, запускается ли pipeline и корректно ли он завершает задачи).
* **Упор:** на автоматизации и надежности процессов. Важно не только запустить CI, но и понять, как он работает (четкая настройка этапов: сборка, тесты, деплой). Этот навык критичен для DevOps.

## Месяц 8: Оркестрация контейнеров (Kubernetes)

* **Фокус:** изучите основы оркестрации контейнеров. Kubernetes и аналогичные системы (Docker Swarm) позволяют масштабировать приложения из контейнеров.
* **Ресурсы:** интерактивные симуляторы (Katacoda с Kubernetes-пособиями), документация Kubernetes («The Kubernetes Book» – бесплатные фрагменты), YouTube-курсы “Kubernetes for Beginners”.
* **Практика:** установите мини-кластер (minikube) и разверните в нём ваши контейнеры. Создайте деплоймент и сервис – проверьте, что приложение доступно через кластер.
* **Оценка:** умение описать манифесты Deployment/Service и запустить их. Успех – приложение функционирует в Kubernetes-кластере. В приоритете понимание архитектуры (pod, node, cluster).
* **Упор:** на концепциях масштабируемости и самовосстановления: узнайте, как Kubernetes заменяет упавшие контейнеры, масштабирует число копий и распределяет нагрузку.

## Месяц 9: Infrastructure as Code (Terraform/Ansible)

* **Фокус:** автоматизируйте конфигурацию инфраструктуры: Terraform или аналогичные инструменты (Ansible, Pulumi). Это позволит описывать и версионировать архитектуру.
* **Ресурсы:** официальный гайд Terraform (hashicorp.com), интерактивные уроки Terraform. Для Ansible – документация и вводные курсы.
* **Практика:** напишите скрипт Terraform для создания серверов/сетей вашей облачной инфраструктуры (напр., в AWS). С помощью кода запустите все ресурсы за один клик и проверьте, что они конфигурируются без ручного вмешательства.
* **Оценка:** если после удаления ресурсов вы можете заново «прогнать» скрипт и получить идентичную инфра — задача решена. Проверьте, что понимаете, как версии кода приводят к изменению инфраструктуры.
* **Упор:** на детальности и повторяемости: каждый элемент (сервер, БД, сеть) должен быть описан кодом. Освойте основы синтаксиса Terraform и команд для планирования/применения конфигурации.

## Месяц 10: Мониторинг, логирование и безопасность

* **Фокус:** настройте мониторинг и логирование приложений. Это важно для отслеживания состояния системы в продакшене. Наиболее популярные инструменты – Prometheus и Grafana. Изучите основы сбора метрик, алертинга, а также базовую безопасность (SSL, настройка прав, OWASP-топ-10).
* **Ресурсы:** туториалы по Prometheus/Grafana, курсы по мониторингу (например, на YouTube: “Prometheus tutorial”), документация (AWS CloudWatch/GCP Monitoring). Для безопасности – статьи OWASP, вводные по шифрованию (SSL/TLS).
* **Практика:** подключите приложение к мониторингу: например, собирайте метрики производительности (загрузка CPU, время ответа). Настройте простой дашборд в Grafana. Внедрите SSL-сертификат (Let’s Encrypt) для своего домена.
* **Оценка:** проверьте, можете ли создавать метрики и алерты (например, срабатывает уведомление при недоступности приложения). Убедитесь, что трафик шифруется (HTTPS) и основные уязвимости закрыты (например, актуальность библиотек).
* **Упор:** на готовности системы к реальной эксплуатации: логирование ошибок, оповещения о сбоях, базовая защита пользовательских данных.

## Месяц 11: Архитектура и совершенствование навыков

* **Фокус:** освоение более высокоуровневых концепций: паттерны архитектуры (MVC, микросервисы), кеширование (Redis, Memcached), очереди сообщений (RabbitMQ, Kafka) для асинхронности. Изучите принципы проектирования (Code Review, TDD, SOLID).
* **Ресурсы:** статьи/книги по системному дизайну (разделы на GeeksforGeeks или Яндекс.Станция, Habr), доклады с конференций. Официальные туториалы Redis/RabbitMQ, YouTube (например, серия “Redis Crash Course”).
* **Практика:** добавьте в своё приложение кеширование запросов (Redis) или очередь задач (Laravel Queue с Redis). Постройте небольшой сервис или микросервис (например, отдельный модуль авторизации), чтобы поэкспериментировать с разделением ответственности.
* **Оценка:** оцените, насколько улучшилось быстродействие/масштабируемость после кеша. Проверьте, умеете ли декомпозировать задачу на сервисы и интегрировать их.
* **Упор:** на чистой архитектуре и качественном коде. Знакомьтесь с чужими проектами на GitHub, чтобы видеть примеры хорошей структуры. Работайте над тем, чтобы писать покрытый тестами код и документировать решения.

## Месяц 12: Итоговый проект и обзор

* **Фокус:** интегрируйте все наработки. Постройте полный проект с React фронтендом и Laravel API на бэкенде, запустите его в Docker/Kubernetes с CI/CD. Обратите внимание на все этапы: от написания кода до деплоя и мониторинга.
* **Ресурсы:** собственные заметки и пройденные курсы. Возможно, найдите контрольный проект на GitHub для вдохновения (например, «Laravel + React boilerplate»).
* **Практика:** реализуйте мини-приложение (блог, таск-менеджер, чат) с фронтендом на React и бэкендом на Laravel, задеплойте всё в облако. Настройте автоматический деплой из репозитория и мониторинг в продакшне.
* **Оценка:** проверьте полноценность решения: работоспособность всех функций, автоматическую сборку и деплой, наличие мониторинга и логирования. Получите отзывы – просьба коллег протестировать приложение.
* **Упор:** на **завершённости** проекта и профессиональном подходе: соблюдении стайлгайдов, тестах, документации. Научитесь объяснять архитектуру своего решения и ловко работать с инструментами, которые вы освоили в течение года.

## 🔹 1. Проверка IP через DNS (nslookup / dig)

Если ты в Windows:
Открой `CMD` и введи:

```bash
nslookup google.com
```

Пример вывода:

```
Name:    google.com
Addresses:  142.250.190.78
```

👉 Это показывает, что DNS-сервер вернул IP-адрес `142.250.190.78` для `google.com`.

## 🔹 2. Проверка маршрута к серверу (tracert или traceroute)

### Windows:

```bash
tracert google.com
```

### Linux/macOS:

```bash
traceroute google.com
```

Пример:

```
1    1 ms     1 ms     1 ms  local-router
2   15 ms    14 ms    15 ms  isp.net
3   38 ms    37 ms    36 ms  core-backbone.net
...
9   96 ms    93 ms    94 ms  google.com [142.250.190.78]
```

👉 **Каждая строка** — это один **"хоп"** (маршрутизатор, через который прошёл пакет).
👉 **Задержка (ms)** показывает, сколько времени занимает прохождение через него.

## 🔹 3. Проверка пинга до сайта (задержка)

```bash
ping google.com
```

Пример:

```
Ответ от 142.250.190.78: число байт=32 время=23мс TTL=117
```

👉 Это проверяет **жив ли сервер**, **время отклика**, и **TTL (time to live)** — сколько хопов осталось.

## 🔹 4. Проверка DNS-записей домена

Если у тебя есть домен и ты хочешь узнать, куда он направлен:

```bash
nslookup mysite.com
```

Или более точно:

```bash
nslookup -type=A mysite.com
```

👉 Это покажет, **какой IP** присвоен домену.

---

## 🔹 5. Типы DNS-записей (что настраивают профи на сервере):

| Тип записи | Что делает                                      |
| ---------- | ----------------------------------------------- |
| A          | Привязывает домен к IPv4 адресу                 |
| AAAA       | Привязывает к IPv6 адресу                       |
| CNAME      | Привязывает один домен к другому                |
| MX         | Почтовый сервер                                 |
| TXT        | Текстовая запись (например, SPF/DKIM для email) |
| NS         | Указывает, какие DNS-сервера отвечают за домен  |

📌 Профи на хостинге через панель или API настраивают DNS-записи при запуске сайта, подключении email и SSL.

---

## 🔹 6. Как профи работают с этим в проектах:

* ⚙️ Покупают домен (Namecheap, GoDaddy, etc.)
* ⚙️ Привязывают его к IP на DNS (через A-запись)
* ⚙️ Настраивают SSL (Let's Encrypt, Cloudflare)
* ⚙️ Тестируют `nslookup`, `traceroute`, `ping`
* ⚙️ Настраивают `firewall`/`Nginx`/`Apache` для правильной маршрутизации
